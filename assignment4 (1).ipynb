# 1
Scrape the details of most viewed videos on YouTube from Wikipedia:
Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
 You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) View

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
rank = []
for i in driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]'):
    rank.append(i.text.replace('/n',' ')) # finding rank of the songs
len(rank)
v = []
for i in driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]'):
    v.append(i.text.replace('/n',' '))  #finding views
date = []
for i in driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]'):
    date.append(i.text.replace('/n',' '))  #finding date of upload
t = []
for i in driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]'):
    t.append(i.text.replace('/n',' '))  # title of the videos
artist=[]
try:
    artists=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in artists:                         #finding artist 
        artist.append(i.text)

except NoSuchElementException:#handling no such element exception
    artist.append('No details available')

except StaleElementReferenceException:#handling Stale element exception
    artist.append('No details available')
```

```python
video_df = pd.DataFrame({})
video_df['Rank'] = rank
video_df['Title'] = t
video_df['Artist'] =  artist
video_df['View(in billions)'] = v
video_df['Upload Date'] = date
video_df
```

# 2
Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
driver.get('https://www.bcci.tv/')
```

```python
navi = driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']")
navi.click()      #clicking on navigation dropdown to get its option
```

```python
inter_fixtures = driver.find_element_by_xpath("//div[@class='drop-down__options']/ul[1]/li[1]/a")
i = inter_fixtures.get_attribute('href')      
driver.get(i) 
#Now after getting navigation option fetch the international feature href attribute and open that page through driver.get()
```

```python
#  Getting paths of title, place,time,date and series of match
title = []
for i in driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']"):
    title.append(i.text.replace('/n',' '))
place = []
for j in driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]/span'):
    place.append(j.text.replace('/n',''))
time = []
for j in driver.find_elements_by_xpath('//span[@class="fixture__time"]'):
    time.append(j.text.replace('/n',''))
date = []
for j in driver.find_elements_by_xpath('//span[@class="fixture__date"]'):
    date.append(j.text.replace('/n',''))
month = []
for j in driver.find_elements_by_xpath('//span[@class="fixture__month"]'):
    month.append(j.text.replace('/n',''))
series = []
for k in driver.find_elements_by_xpath('//div[@class="fixture__format-strip"]'):
    series.append(k.text.replace('/n',''))
```

```python
# The series that fetched were in format of 'TEST ICC WORLD \n TEST CHAMPIONSHIP' to remove '\n' from the list I used this code
series_n = []
for s in series:
    new_string = s.replace("\n", " ")
    series_n.append(new_string)
```

```python
#For date month and date were fetched seperately to map it in single list , used map and lambda function
match_date = []
match_date = list(map(lambda x, y: x+ ' ' +y, date, month))
```

```python
international_fixtures = pd.DataFrame({})
international_fixtures['title'] = title
international_fixtures['Place'] = place
international_fixtures['Date'] = match_date
international_fixtures['Time'] = time
international_fixtures['series'] = series_n
international_fixtures
```

```python
driver.get('https://github.com/')
```

```python
trending = driver.find_element_by_xpath('//ul[@class="list-style-none mb-3"][2]/li[3]/a')
driver.get(trending.get_attribute('href'))
#home page to trending page find the path of trending from explore dropbox then after fetching its href I used get method.
```

```python
url=[]
for i in driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]/a'):
    url.append(i.get_attribute('href'))
#Fetching urls for individual trending repositories
```

```python
title =[]
descp =[]
c =[]
lan=[]
#finding paths from each url and used exception handling for absence of any element
for i in url:  
    driver.get(i)
    time.sleep(2)
    try:
        p=driver.find_element_by_xpath('//h1[@class=" d-flex flex-wrap flex-items-center break-word f3 text-normal"]')
        title.append(p.text.replace('/n',''))
    except NoSuchElementException:
        title.append('-')
    try:
        p=driver.find_element_by_xpath('//div[@class="BorderGrid-cell"]/p')
        descp.append(p.text.replace('/n',''))
    except NoSuchElementException:
        descp.append('-')
    try:
        p=driver.find_element_by_xpath('//div[@class="BorderGrid BorderGrid--spacious"]/div[4]/div/h2/a/span')
        c.append(p.text.replace('/n',''))
    except NoSuchElementException:
        c.append('-')
    try:
        p=driver.find_element_by_xpath('//div[@class="BorderGrid BorderGrid--spacious"]/div[5]/div/ul')
        lan.append(p.text.replace('/n',''))
    except NoSuchElementException:
        lan.append('-')
```

```python
#titles and languages had '\n' in records to remove it used this code for both
title_n = []
for s in title:
    new_string = s.replace("\n", " ")
    title_n.append(new_string)
```

```python
lang_n = []
for s in lan:
    new_string = s.replace("\n", " ")
    lang_n.append(new_string)
```

```python
github_repost = pd.DataFrame({})
github_repost['Title'] = title_n
github_repost['Language'] = lang_n
github_repost['contributers number'] = c
github_repost['Description'] = descp
github_repost
```

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
view_all = driver.find_element_by_xpath('//tbody/tr/td[2]/span[@class="whitetext"][2]/a')
driver.get(view_all.get_attribute('href'))  
#from home page to view all datesets page fetch the href of that page and open through driver.get
```

```python
dt=[]
for i in driver.find_elements_by_xpath('//tbody/tr/td[2]/p[@class="normal"]'):
    dt.append(i.text.replace('/n',''))
```

```python
#In dt both name and datatypes were fetched at alternate place at even place name were there and at odd datatype
#So extracted both by this code in different list
c=0
name=[]
datatype=[]
for i in dt:
    if(c%2==0):
        name.append(i)
        c=c+1
    elif(c%2==1):
        datatype.append(i)
        c=c+1
```

```python
#fetched remaining 5 columns from the table by finding the paths
task=[]
for i in driver.find_elements_by_xpath('//tbody/tr/td[3]/p[@class="normal"]'):
    task.append(i.text.replace('/n',''))
no_insta=[]
for i in driver.find_elements_by_xpath('//tbody/tr/td[5]/p[@class="normal"]'):
    no_insta.append(i.text.replace('/n',''))
attri=[]
for i in driver.find_elements_by_xpath('//tbody/tr/td[4]/p[@class="normal"]'):
    attri.append(i.text.replace('/n',''))
no_attri=[]
for i in driver.find_elements_by_xpath('//tbody/tr/td[6]/p[@class="normal"]'):
    no_attri.append(i.text.replace('/n',''))
yr=[]
for i in driver.find_elements_by_xpath('//tbody/tr/td[7]/p[@class="normal"]'):
    yr.append(i.text.replace('/n',''))
```

```python
datasets_uci = pd.DataFrame({})
datasets_uci['Dataset name'] = name
datasets_uci[' Data type'] = datatype
datasets_uci['Task'] = task
datasets_uci['Attribute type'] = attri
datasets_uci['No of instances'] = no_insta
datasets_uci[' No of attribute'] = no_attri
datasets_uci['Year'] = yr
datasets_uci
```
