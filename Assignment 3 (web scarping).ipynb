4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on 
www.flipkart.com and scrape following details for all the search results displayed on 1st page. 
Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, 
“Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display 
Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. 
Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe 
and CSV.


```python
driver.get('http://www.flipkart.com/')
```

```python
search_product = driver.find_element_by_xpath('//input[@class="_3704LK"]')
search_product.send_keys('pixel 4A')
```

```python
search_click = driver.find_element_by_xpath('//button[@class="L0Z3Pu"]')
search_click.click()
```

```python
mobile_name = []
for j in driver.find_elements_by_xpath('//div[@class="_4rR01T"]'):
    mobile_name.append(j.text.replace('/n',''))
```

```python
urls=[]
for i in driver.find_elements_by_xpath('//a[@class="_1fQZEK"]'):
    urls.append(i.get_attribute('href'))
```

```python
prices = []
for i in urls:
    driver.get(i)
    time.sleep(2)
    try:
        r=driver.find_element_by_xpath("//div[@class='_30jeq3 _16Jk6d']")
        prices.append(r.text.replace('/n',''))
    except NoSuchElementException:
        prices.append('-')
```

```python
ram_rom = []
battery = []
cam = []
display =[]
processor = []
for i in urls:
    driver.get(i)
    time.sleep(2)
    try:
        r=driver.find_element_by_xpath("//div[@class='_2418kt']/ul/li[1]")
        ram_rom.append(r.text.replace('/n',''))
    except NoSuchElementException:
        ram_rom.append('-')
    try:
        r=driver.find_element_by_xpath("//div[@class='_2418kt']/ul/li[2]")
        display.append(r.text.replace('/n',''))
    except NoSuchElementException:
        display.append('-')
    try:
        r=driver.find_element_by_xpath("//div[@class='_2418kt']/ul/li[3]")
        cam.append(r.text.replace('/n',''))
    except NoSuchElementException:
        cam.append('-')
    try:
        r=driver.find_element_by_xpath("//div[@class='_2418kt']/ul/li[4]")
        battery.append(r.text.replace('/n',''))
    except NoSuchElementException:
        battery.append('-')
    try:
        r=driver.find_element_by_xpath("//div[@class='_2418kt']/ul/li[5]")
        processor.append(r.text.replace('/n',''))
    except NoSuchElementException:
        processor.append('-')
```

```python
RAM_ = []
for i in ram_rom:
        j= ''                            #seperating ram and rom values
        j = i.split("|")
        RAM_.append(j)
RAM = []
ROM = []
c=2                    #inserting ram and rom values in individual lists
for i in RAM_:
    for j in range(2):
        if(c%2==0):
            RAM.append(i[0])
            c=c+1
        else:
            ROM.append(i[1])
            c=c+1
dis = []
for i in display:     #seperating display size and resolution
        j= ''
        j = i.split(")")
        dis.append(j)
dis_size = []
dis_resolution = []
c=2
for i in dis:
    for j in range(2):    # inserting display size and resolution in seperate lists
        if(c%2==0):
            dis_size.append(i[0])
            c=c+1
        else:
            dis_resolution.append(i[1])
            c=c+1
name = []
for i in mobile_name:    #seperating mobile name from product title
        j= ''
        j = i.split("(")
        name.append(j)
colors = []
c=0
for i in name:
    colors.append(name[c][1])   # seperating color from the product title name
    c+=1
mobile_colors = []
for i in colors:
        j= ''
        j = i.split(",")
        mobile_colors.append(j)
col_mobile = []
c=0
for j in mobile_colors:
    col_mobile.append(mobile_colors[c][0])
    c+=1
mobilename = []
c=0
for j in name:
    mobilename.append(name[c][0])
    c+=1
cam_l = []
for i in cam:                # seperating primary camera 
        j= ''
        j = i.split("|")
        cam_l.append(j)
cam_primary_=[]
c=0
for i in cam_l:
    cam_primary_.append(cam_l[c][0])
    c+=1
```

```python
mobiles_df = pd.DataFrame({})
```

```python
mobiles_df['mobilename'] = mobilename
mobiles_df['color'] = col_mobile
mobiles_df['processor'] = processor
mobiles_df['display resolution'] = dis_resolution
mobiles_df['display size'] = dis_size
mobiles_df['battery'] = battery
mobiles_df['RAM'] = RAM
mobiles_df['ROM'] = ROM
mobiles_df['camera'] = cam
mobiles_df['Prices'] = prices
mobiles_df['urls'] = urls
mobiles_df['primary_cam'] = cam_primary_
```

7. Write a program to scrap all the available details of top 10 gaming laptops from digit.in.

```python
driver.get('https://www.digit.in/search?keyword=gaming+laptops')
```

```python
product_name = []
for i in driver.find_elements_by_xpath("//div[@class='searchProduct-desc']"):
    product_name.append(i.text.replace('/n',''))
```

```python
urls=[]
for i in driver.find_elements_by_xpath('//div[@class="searchPage"]/a')[:10]:
    urls.append(i.get_attribute('href'))
```

```python
rating = []
prices = []
memory = []
processor = []
os = []
display =[]
for i in urls[:10]:
    driver.get(i)
    time.sleep(3)
    try:
        r=driver.find_element_by_xpath("//div[@class='score-Bar']/ul/li/p")
        rating.append(r.text.replace('/n',''))
    except NoSuchElementException:
        rating.append('-')
    try:
        r=driver.find_element_by_xpath("//div[@class='Specs-Wrap  notspecs ']/ul/li[4]/div")
        memory.append(r.text.replace('/n',''))
    except NoSuchElementException:
        memory.append('-')
    try:
        r=driver.find_element_by_xpath("//div[@class='Specs-Wrap  notspecs ']/ul/li[3]/div")
        processor.append(r.text.replace('/n',''))
    except NoSuchElementException:
        processor.append('-')
    try:
        r=driver.find_element_by_xpath("//div[@class='Specs-Wrap  notspecs ']/ul/li[1]/div")
        os.append(r.text.replace('/n',''))
    except NoSuchElementException:
        os.append('-')
    try:
        r=driver.find_element_by_xpath("//div[@class='Specs-Wrap  notspecs ']/ul/li[2]/div")
        display.append(r.text.replace('/n',''))
    except NoSuchElementException:
        display.append('-')
    try:
        r=driver.find_element_by_xpath("//div[@class='Block-price']/b")
        prices.append(r.text.replace('/n',''))
    except NoSuchElementException:
        prices.append('-')
```

```python
official_web = []
market_sta = []
rel_dat = []
for i in urls[:10]:
    driver.get(i)
    time.sleep(3)
    try:
        r=driver.find_element_by_xpath("//div[@class='status-bar only-desktop']/div[3]/b")
        official_web.append(r.text.replace('/n',''))
    except NoSuchElementException:
        official_web.append('-')
    try:
        r=driver.find_element_by_xpath("//div[@class='status-bar only-desktop']/div[1]/b")
        market_sta.append(r.text.replace('/n',''))
    except NoSuchElementException:
        market_sta.append('-')
    try:
        r=driver.find_element_by_xpath("//div[@class='status-bar only-desktop']/div[2]/b")
        rel_dat.append(r.text.replace('/n',''))
    except NoSuchElementException:
        rel_dat.append('-')
    
```

```python
laptop_df = pd.DataFrame({})
laptop_df["name"] = product_name
laptop_df["price"] = prices
laptop_df["url"] = urls
laptop_df["memory"] = memory
laptop_df["os"] = os
laptop_df["proceesor"] = processor
laptop_df["display"] = display
laptop_df["rating"] = rating
laptop_df["official web"] = official_web
laptop_df["market_status"] = market_sta
laptop_df["relaease date"] =rel_dat
laptop_df
```
