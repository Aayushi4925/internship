# 1.Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’

```python
header_tags = pd.DataFrame({})
h1_n  =pd.DataFrame({})
def wiki(l):
    pg = requests.get(l)
    soup  = BeautifulSoup(pg.content)
    h1_tag = soup.find_all('h1',class_='firstHeading')
    h2_tag = soup.find_all('h2',class_='mp-h2')
    h1 = []
    h2 = []
    for i in h1_tag:
        h1.append(i.text.replace('\n',''))
    for i in h2_tag:
        h2.append(i.text.replace('\n',''))
    h1_n['h1'] = h1
    header_tags['header_tags_2'] = h2
    
    
```

# 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com

```python
books_df = pd.DataFrame({})
def books(a):
    page_3 = requests.get(a)
    soup = BeautifulSoup(page_3.content)
    books_name = soup.find_all('h4',class_='italic')
    book_author = soup.find_all('p',class_='sans bold')
    book_genre = soup.find_all('p',class_='genre-links hidden-phone')
    book_review = soup.find_all('p',class_='excerpt')
    bookname = []
    bookauthor = []
    bookgenre = []
    bookreview = []
    for i in books_name[0:5]:
        bookname.append(i.text.replace('\n',''))
    for i in book_author[0:5]:
        bookauthor.append(i.text.replace('\n',''))
    for i in book_genre[0:5]:
        bookgenre.append(i.text.replace('\n',''))
    for i in book_review[0:5]:
        bookreview.append(i.text.replace('\n',''))
    books_df['book_name'] = bookname
    books_df['book_author'] = bookauthor
    books_df['books_genre']  = bookgenre
    books_df['review'] = bookreview
   
    
```

```python
books('https://bookpage.com/reviews?book_genre=nonfiction&page=1')
```

# 7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
mobile_df = pd.DataFrame({})
```

```python
def mobile_ph(c):
    page_5 = requests.get(c)
    soup = BeautifulSoup(page_5.content)
    mobile_name = soup.find_all('span',class_='a-size-medium a-color-base a-text-normal')
    mobile_price = soup.find_all('span',class_='a-whole-price')
    rating = soup.find_all('div',class_='a-row a-size-small')
    img = soup.find_all('img',class_='s-image')
    ratings =[]
    mobilename = []
    mobileprice = []
    imgs = []
    for i in mobile_name:
        mobilename.append(i.text.replace('\n',''))
    for i in mobile_price:
        mobileprice.append(i.text.replace('\n',''))
    for i in rating:
        ratings.append(i.text.replace('\n',''))
    for i in img:
        imgs.append(i.get('src'))
    mobile_df['mobile_name'] = mobilename
    mobile_df['mobile_price'] = mobileprice
    mobile_df['ratings'] = ratings
    mobile_df['url_imgs'] = imgs
    
    
    
    
    
```

# 8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
forecast_df = pd.DataFrame({})
```

```python
def forecast(l):
    pg = requests.get(l)
    soup  = BeautifulSoup(pg.content)
    period = []
    for i in soup.find_all('p',class_='period-name'):
        period.append(i.text.replace('\n',''))
    period = period[:7]
    Temp=[]
    for i in soup.find_all('p',class_="temp"):
        Temp.append(i.text.replace('\n',''))
    Temp=Temp[:7]
    short_desc = []
    for i in soup.find_all('p',class_='short-desc'):
        short_desc.append(i.text.replace('\n',''))
    short_desc = short_desc[:7]
    detail_desc = []
    for i in soup.find_all('div',class_='col-sm-10 forecast-text'):
        detail_desc.append(i.text.replace('\n',''))
    detail_desc = detail_desc[:7]
    forecast_df['period'] = period
    forecast_df['temp'] = Temp
    forecast_df['short_desc'] = short_desc
    forecast_df['description'] = detail_desc
```

```python
forecast('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YJeZdrUzZPY')
```
